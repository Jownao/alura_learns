```python
# Fazendo função que separa palavras somente se for Alfabético

def separa_palavras(lista_tokens):
  lista_palavras = []
  # Para cada token na lista_tokens
  for token in lista_tokens:
    # Se a palavra for toda alfabética
    if token.isalpha():
      # Adiciona na lista de palavras
      lista_palavras.append(token)
  return lista_palavras

```

## Normalização

```python
# Quando contamos tipos diferentes de palavras temos por exemplo 
# "Exemplo" e "exemplo" como tipos diferentes sendo que são as mesmas palavras, 
# para isso vamos criar uma função para normalizar todas essas palavras em minúsculas

def normalizacao (lista_palavras):
  lista_normalizada = []
  for palavra in lista_palavras:
    lista_normalizada.append(palavra.lower())
  return lista_normalizada

lista_normalizada = normalizacao(lista_palavras)
```

### Inserindo letras

```python
lista = "lgica"
(lista[:1],lista[1:])
```

```python
palavra_exemplo = 'lgica'

def gerador_palavras(palavra):
  fatias = []
  for i in range(len(palavra)+1):
    fatias.append((palavra[:i],palavra[i:]))

  #print(fatias)
  palavras_geradas = insere_letras(fatias)
  return palavras_geradas


def insere_letras(fatias):
  novas_palavras = []
  letras = 'abcdefghijklmnopqrstuvwxyzàáâãèéêìíîòóôõùúûç'
  for E,D in fatias:
    for letra in letras:
      novas_palavras.append(E+ letra + D)
  return novas_palavras


palavras_geradas = gerador_palavras(palavra_exemplo)
print(palavras_geradas)
```

### Corretor , avaliador e base de teste

```python
def corretor(palavra):
  palavras_geradas = gerador_palavras(palavra)
  palavra_correta = max(palavras_geradas, key=probabilidade)
  return palavra_correta
```

```python
frequencia = nltk.FreqDist(lista_normalizada)
frequencia.most_common(10)

total_palavras = len(lista_normalizada)
```

```python
# Função para calcular a probabilidade de uma palavra específica dentro da nossa base
def probabilidade(palavra_gerada):
  # Retornar a frequencia passando a palavra como parâmetro e dividindo pelo total de palavra , dando como resultado a probabilidade da mesma
  return frequencia[palavra_gerada]/total_palavras
```

```python
def cria_dados_teste(nome_arquivo):
    lista_palavras_teste = []
    f = open(nome_arquivo, "r")
    for linha in f:
        correta, errada = linha.split()
        lista_palavras_teste.append((correta, errada))
    f.close()
    return lista_palavras_teste

lista_teste = cria_dados_teste("/content/drive/MyDrive/Colab Notebooks/Datascience/Corretor Ortografico - NLP/palavras.txt")
lista_teste
```

```python
def avaliador(testes):
  numero_palavras=len(testes)
  acertou = 0
  for correta,errada in testes:
    palavra_corrigida = corretor(errada)
    if palavra_corrigida == correta:
      acertou+=1
  taxa_acerto = (acertou*100)/numero_palavras
  print(f'Taxa de acerto: {round(taxa_acerto,2)}% de {numero_palavras} palavras')

avaliador(lista_teste)
```

### Deletando caractere

```python
# Novo Algoritmo

def deleta_caractere(fatias):
  novas_palavras = []
  for E,D in fatias:
      novas_palavras.append(E+ D[1:])
  return novas_palavras
```

### Invertendo caractere


```python
# Novo algoritmo

def inverte_caractere(fatias):
  novas_palavras = []
  for E,D in fatias:
    if len(D) > 1:
      novas_palavras.append(E+ D[1] + D[0] + D[2:])
  return novas_palavras
```

```python
# Novo Gerador

def gerador_palavras(palavra):
  fatias = []
  for i in range(len(palavra)+1):
    fatias.append((palavra[:i],palavra[i:]))

  #print(fatias)
  palavras_geradas = insere_letras(fatias)
  palavras_geradas += deleta_caractere(fatias)
  palavras_geradas += troca_caractere(fatias)
  palavras_geradas += inverte_caractere(fatias)
  return palavras_geradas
```

### Melhorando o avaliador de palavras

```python
def avaliador(testes,vocabulario):
  numero_palavras=len(testes)
  acertou = 0
  desconhecida = 0
  for correta,errada in testes:
    palavra_corrigida = corretor(errada)
    if palavra_corrigida == correta:
      acertou+=1
    else:
      desconhecida += (correta not in vocabulario)
  taxa_acerto = (acertou*100)/numero_palavras
  taxa_desconhecida = (desconhecida*100)/numero_palavras
  print(f'Taxa de acerto: {round(taxa_acerto,2)}% de {numero_palavras} palavras')
  print(f'Taxa de palavras desconhecidas: {round(taxa_desconhecida,2)}% de {numero_palavras} palavras')

vocabulario = set(lista_normalizada)
avaliador(lista_teste,vocabulario)
```

### Gerador melhorado

```python
# Gerador com outro gerador dentro para corrigir erro de duas palavras repetidas como "alaaface" (alface)

palavra = 'alaaface'

def gerador_melhorado(palavras_geradas):
  novas_palavras = []
  for palavra in palavras_geradas:
    novas_palavras += gerador_palavras(palavra)
  return novas_palavras

teste = gerador_melhorado(gerador_palavras(palavra))

# Teste para ver se ele ta gerando a palavra correta
"alface" in teste 
```

### Melhorando corretor

```python
def novo_corretor(palavra):
    palavras_geradas = gerador_palavras(palavra)
    palavras_turbinado = gerador_melhorado(palavras_geradas)
    todas_palavras = set(palavras_geradas + palavras_turbinado)
    candidatos = [palavra]
    for palavra in todas_palavras:
        if palavra in vocabulario:
            candidatos.append(palavra)
    #print(len(candidatos))
    palavra_correta = max(candidatos, key=probabilidade)
    return palavra_correta

novo_corretor('loggica')
```

### Avaliando resultado dos dois corretores

```python
def avaliador(testes,vocabulario):
  numero_palavras=len(testes)
  acertou = 0
  desconhecida = 0
  print('Errada │ COR 1 │ COR 2')
  for correta,errada in testes:
    palavra_corrigida = novo_corretor(errada)
    desconhecida += (correta not in vocabulario)
    if palavra_corrigida == correta:
      acertou+=1
    else:
        print(errada + " - " + corretor(errada) + " - " + palavra_corrigida)
  taxa_acerto = (acertou*100)/numero_palavras
  taxa_desconhecida = (desconhecida*100)/numero_palavras
  print(f'Taxa de acerto: {round(taxa_acerto,2)}% de {numero_palavras} palavras')
  print(f'Taxa de palavras desconhecidas: {round(taxa_desconhecida,2)}% de {numero_palavras} palavras')

vocabulario = set(lista_normalizada)
avaliador(lista_teste,vocabulario)
```

* Pelo fato do segundo corretor fazer a correção até duas casa de erro, ele gera uma palavra diferente da que precisamos, e se a mesma for gerada a probabilidade pode acabar sendo menor fazendo com que a de maior seja escolhida, como podemos ver acontecendo acima.
